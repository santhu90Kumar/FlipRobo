```python
import pandas
import selenium
from selenium import webdriver
import time
import requests
import pandas as pd
```

```python
url1="https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001"
```

```python
driver1=webdriver.Chrome("chromedriver.exe")
driver1.get(url1)
```

```python
City=[400001,110001,122001,411001,560001,500001,600001,700001,380001,390001,462001,134101,121001,201001,450001,302001,682001,
     226001,141001,250001,563102,440001,422001,132101,800001,360001,124001,131001,394101,140604]
```

```python
Url=[]
for i in City:
    Url.append(f'https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId={i}')
Url
```

```python
KM=[]
Owner=[]
Fuel=[]
Description=[]
Price=[]
Transmission=[]
Variant=[]
Description1=[]
City=[]


for i in Url:
    driver1.get(i)
    driver1.execute_script("window.scrollTo(0, 12000000);")
    time.sleep(25)
    
    
    city=driver1.find_elements_by_xpath("//p[@class='_1OR7d']")        
    km=driver1.find_elements_by_xpath("//ul[@class='bVR0c']//li[1]")
    for i in km:
        for j in city:
            KM.append(i.text)
            City.append(j.text)
        
    owner=driver1.find_elements_by_xpath("//ul[@class='bVR0c']//li[2]")
    for i in owner:
        Owner.append(i.text)
        
    fuel=driver1.find_elements_by_xpath("//ul[@class='bVR0c']//li[3]")
    for i in fuel:
        Fuel.append(i.text)
        
    desc=driver1.find_elements_by_xpath("//h2[@class='_3FpCg']")
    for i in desc:
        Description.append(i.text)
        
    price=driver1.find_elements_by_xpath("//div[@class='_7udZZ']")
    for i in price:
        Price.append(i.text)
    
    trans=driver1.find_elements_by_xpath("//p[@class='cvakB']/span")
    for i in trans:
        Transmission.append(i.text)
        
    variant=driver1.find_elements_by_xpath("//div[@class='_1jpRU']/p")
    for i in variant:
        Variant.append(i.text)
        
    desc1=driver1.find_elements_by_xpath("//ul[@class='bVR0c']")
    for i in desc1:
        Description1.append(i.text)
        
Year=[i.split(' ',2)[0] for i in Description]
Brand_Name=[i.split(' ',2)[1] for i in Description]
Model=[i.split(' ',2)[2] for i in Description]       
        

        
print('Length of Km',len(KM))
print('Length of Owner',len(Owner))
print('Length Of Fuel',len(Fuel))
print('Length Of Description',len(Description))
print('Length Of Price',len(Price))
print('Length Of Transmission',len(Transmission))
print('Length Of Variant',len(Variant))
print('Length Of Year',len(Year))
print('Length Of Brand_Name',len(Brand_Name))
print('Length Of Model',len(Model))
print('Length Of Description1',len(Description1))
```

```python
Car=pd.DataFrame()

Car['year']=Year
Car['company name']=Brand_Name
Car['car model']=Model
Car['km driven']=KM1
Car['ownership']=Owner
Car['Fuel']=Fuel
Car['Price']=Price
Car['Transmission']=Transmission
Car['Variant']=Variant
Car['City']=City
Car
```
